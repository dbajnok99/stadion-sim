# Evaluation of Task 5 impatient fans

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## No ultras

```python
df = pd.read_csv("impatient_no_overload.csv")
```

```python
df
```

```python
avg_wait_time = df['Avg Wait (s)'].mean()
avg_switched_wait = df['Avg Wait (Switched)'].mean()
avg_not_switched_wait = df['Avg Wait (Static)'].mean()
avg_lane_changes = df['Lane Changes'].mean()
avg_inside_by_kickoff = df['Inside by Kickoff (%)'].mean()
avg_expected_fans = df['Expected Fans'].mean()

print("Average Wait Time:", avg_wait_time )
print("Average Switched Wait:", avg_switched_wait)
print("Average Not Switched Wait:", avg_not_switched_wait)
print("Average Lane Changes:", avg_lane_changes)
print("Average Inside by Kickoff:", avg_inside_by_kickoff)
print("Average Expected Fans:", avg_expected_fans)
```

```python
small_sample = df.sample(n=10)
small_sample
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

color_switched = '#ef4444' # Red
color_static = '#3b82f6'   # Blue
color_avg = 'gray'

line1, = ax1.plot(small_sample['Run ID'], small_sample['Avg Wait (Switched)'], label='Avg Wait (Switched)', marker='o', linestyle='-', color=color_switched, linewidth=2)
line2, = ax1.plot(small_sample['Run ID'], small_sample['Avg Wait (Static)'], label='Avg Wait (Static)', marker='o', linestyle='-', color=color_static, linewidth=2)
line3, = ax1.plot(small_sample['Run ID'], small_sample['Avg Wait (s)'], label='Overall Avg Wait', marker='', linestyle=':', color=color_avg, alpha=0.5)

ax1.set_xlabel('Run ID', fontsize=12)
ax1.set_ylabel('Average Wait Time (seconds)', fontsize=12)
ax1.tick_params(axis='y')
ax1.grid(True, alpha=0.3)
ax2 = ax1.twinx()
color_changers = '#10b981'
bar1 = ax2.bar(small_sample['Run ID'], small_sample['Lane Changes'], label='Nr. Lane Changers', color=color_changers, alpha=0.3, width=0.4)

ax2.set_ylabel('Number of Lane Changers', fontsize=12, color=color_changers)
ax2.tick_params(axis='y', labelcolor=color_changers)
lines = [line1, line2, bar1]
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

plt.title('Wait Times Breakdown & Lane Change Volume', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

```

```python
changers = avg_lane_changes
non_changers = avg_expected_fans - avg_lane_changes

labels = ['Changed Lane', 'Did Not Change']
sizes = [changers, non_changers]
colors = ['#ef4444', '#3b82f6']
explode = (0.1, 0)
plt.figure(figsize=(10, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140, textprops={'fontsize': 14})

plt.title(f'Proportion of Lane Changers\n(Average across {len(df)} runs)', fontsize=16)
plt.axis('equal')
```
